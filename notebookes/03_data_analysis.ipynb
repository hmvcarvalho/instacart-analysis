```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the data to variables
df_instacart_orders = pd.read_csv('../data/data_clean/instacart_orders_clean.csv')
df_products = pd.read_csv('../data/data_clean/products_clean.csv')
df_order_products = pd.read_csv('../data/data_clean/order_products_clean.csv')
df_aisles = pd.read_csv('../data/raw/aisles.csv', sep=';')
df_departments = pd.read_csv('../data/raw/departments.csv', sep=';')
```

```python
# verify 'order_hour_of_day' and 'order_dow' min, max and num unique values
df_instacart_orders.agg({'order_hour_of_day' : ['min','max', 'nunique'],
                         'order_dow' : ['min','max', 'nunique']                      
                         })
```

- values in columns make totaly sense as they have 24 and 7 unique in the ranges between 0 and 24/6.

```python
# plot to show how many people make orders at each hour of the day

# count users per hour
user_per_hour = df_instacart_orders.groupby('order_hour_of_day')['user_id'].nunique()

# Create bar chart
plt.figure(figsize=(10, 6))
plt.bar(user_per_hour.index, user_per_hour.values)
plt.title('Number of Orders by Hour of Day')
plt.xlabel('Hour of Day (0-23)')
plt.ylabel('Number of Users')
plt.xticks(range(0, 24))
plt.grid(axis='y', alpha=0.3)
plt.show()
```
